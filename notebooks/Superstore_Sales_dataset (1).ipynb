```python
#Sales distribution by year
sales['Year']=sales['Order Date'].dt.year
sales_by_year=round(sales.groupby('Year')['Sales'].sum())
plt.figure(figsize=(7,6))
sales_by_year.plot(kind='line', marker='o')
plt.title('Sales distribution by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales ($)')
plt.xticks(sales_by_year.index.astype(int))
plt.tight_layout()
for year, sale in sales_by_year.items():
    plt.text(year, sale +5000, str(int(sale)), ha='center', va='bottom')
plt.show()
```

```python
#Yearly Order Count
yearly_order_count=sales.groupby('Year')['Order ID'].nunique()
yearly_order_count.plot(kind='line', marker='s',figsize=(7,6))
plt.title('Yearly Order Count')
plt.xlabel('Year')
plt.ylabel('Order Count')
plt.xticks(yearly_order_count.index.astype(int))
plt.tight_layout()
for year, count in yearly_order_count.items():
    plt.text(year, count +5, str(count), ha='right', va='bottom')
plt.show()
```

```python
#Monthly Sales by Year
sales['Year'] = sales['Order Date'].dt.year
sales['Month'] = sales['Order Date'].dt.month
monthly_sales = sales.groupby(['Year', 'Month'])['Sales'].sum().reset_index()
pivot_table = monthly_sales.pivot(index='Month', columns='Year', values='Sales')
pivot_table.plot(kind='line', marker='o', figsize=(8,6))
plt.title('Monthly Sales by Year')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
#Seasonal Sales
sales['Month']=sales['Order Date'].dt.month
def get_season(month):
    if month in [12,1,2]:
        return 'Winter'
    elif month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    else:
        return 'Fall'

sales['Season']=sales['Month'].apply(get_season)
seasonal_sales=sales.groupby(['Year','Season'])['Sales'].sum().unstack()
seasonal_sales.plot(kind='line', title='Seasonal Sales', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales ($)')
plt.xticks(seasonal_sales.index.astype(int))
plt.legend(title='Season')
plt.tight_layout()
plt.show()
```

```python
#Regional Sales and Their Percentage Share
region_sales=sales.groupby('Region')['Sales'].sum().round(2)
region_sales_percent=round((region_sales/region_sales.sum())*100,2).astype(str)+'%'
result=pd.concat([region_sales,region_sales_percent],axis=1)
result.columns=['Sales','Percentage']
result.sort_values('Sales', ascending=False)
```

```python
#Regional Sales Percentage Distribution
sales['Year']=sales['Order Date'].dt.year
region_sales_percent=round((region_sales/region_sales.sum())*100,2).sort_values(ascending=False)
plt.figure(figsize=(6,5))
bars = plt.bar(region_sales_percent.index, region_sales_percent.values)
plt.title('Regional Sales Percentage Distribution')
plt.xlabel('Region')
plt.ylabel('Sales (%)')
plt.xticks(rotation=0)
plt.tight_layout()

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.1, f'{height}%',
             ha='center', va='bottom')

plt.show()
```

```python
#Regional trend by year
region_year= sales.groupby(['Year','Region'])['Sales'].sum().sort_values(ascending=False).unstack()
region_year.plot(kind='line', marker='o', figsize=(7,6))
plt.title('Regional Trend by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales ($)')
plt.xticks(region_year.index.astype(int))
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

```python
#Top 10 States by Number of Orders
sales['State'].value_counts().head(10)
```

```python
#Top 10 States by Sales and Their Percentage
state_sales=sales.groupby('State')['Sales'].sum().round()
state_sales_percent=round((state_sales/state_sales.sum())*100,2).astype(str)+'%'
result=pd.concat([state_sales,state_sales_percent],axis=1)
result.columns=['Sales','Percentage']
result.sort_values('Sales', ascending=False).head(10)
```

```python
#Top 10 States with Highest Sales
top_states=sales.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_states.plot(kind='bar', color='orange')
plt.title('Top 10 States with Highest Sales')
plt.xlabel('State')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top 10 cities by Number of Orders
sales['City'].value_counts().head(10)
```

```python
#Most Active Cities by Order Count
sales_city=sales['City'].value_counts().head(10)
sns.barplot(x=sales_city.index, y=sales_city.values)
plt.title('Top 10 Cities by Number of Orders')
plt.xlabel('City')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
```

```python
#Top 10 Cities by Sales Amount
top_cities=sales.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_cities.plot(kind='bar', color='blue')
plt.title('Top 10 Cities by Sales Amount ($)')
plt.xlabel('City')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Category-wise Sales and Their Percentage of Total Sales
category_sales=sales.groupby('Category')['Sales'].sum().round(1)
category_sales_percent=round((category_sales/category_sales.sum())*100,2).round(1).astype(str)+'%'
result=pd.concat([category_sales,category_sales_percent],axis=1)
result.columns=['Sales','Percentage']
result.sort_values('Sales', ascending=False)
```

```python
#Sales distribution by category
category_sales=sales.groupby('Category')['Sales'].sum()
plt.figure(figsize=(6, 4))
category_sales.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    ylabel='' )
plt.title('Sales by Category')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
#Top 10 Sub-categories by Number of Orders
sales['Sub-Category'].value_counts().head(10)
```

```python
#Sales by Sub-Categories and Their Percentage of Total Sales
sub_category_sales=sales.groupby('Sub-Category')['Sales'].sum().round(1)
sub_category_sales_percent=round((sub_category_sales/sub_category_sales.sum())*100,2).astype(str)+'%'
result=pd.concat([sub_category_sales,sub_category_sales_percent],axis=1)
result.columns=['Total Sales','Percentage']
result.sort_values('Total Sales',ascending=False).head(10)
```

```python
#Top 10 Highest Revenue Sub-Categories
top_subcategories=sales.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_subcategories.plot(kind='bar', color='red')
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top-10 Products by Total Revenue
sales.groupby(['Product Name','Sub-Category']).agg({'Sales':['sum','count']}).sort_values(('Sales','sum'),ascending=False).head(10)
```

```python
#Top 10 Most Frequently Sold Products by Sub-Category
sales.groupby(['Product Name','Sub-Category']).agg({'Sales':['sum','count']}).sort_values(('Sales','count'),ascending=False).head(10)
```

```python
#Sales and Order Distribution by Customer Segment
sales_segment = sales['Segment'].value_counts()
sales_segment_percentage = sales_segment / sales_segment.sum() * 100
sales_segment_percentage_str = sales_segment_percentage.round(1).astype(str) + '%'
sales_segment_USD = sales.groupby('Segment')['Sales'].sum().round(1)

result = pd.concat([sales_segment, sales_segment_percentage_str, sales_segment_USD], axis=1)
result.columns = ['Number of Orders', ' Sales Percentage', 'Total Sales ($)']
result.sort_values('Number of Orders', ascending=False).head(10)
```

```python
#Distribution of Segments
segment_counts=sales['Segment'].value_counts()
plt.figure(figsize=(6,4))
segment_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightyellow'])
plt.title('Distribution of Segments')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#Sales distribution by Ship Mode
sales_shipmode=sales['Ship Mode'].value_counts()
shipmode_sales= round(sales.groupby('Ship Mode')['Sales'].sum()*100/sales['Sales'].sum(),1).astype(str)+'%'
result=pd.concat([sales_shipmode,shipmode_sales],axis=1)
result.columns=['Number of Orders',' Sales Percentage']
result.sort_values('Number of Orders',ascending=False)
```

```python
#Percentage of Sales by Shipping Type
shipmode_sales= round(sales.groupby('Ship Mode')['Sales'].sum()*100/sales['Sales'].sum(),1)
plt.figure(figsize=(6,4))
shipmode_sales.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightyellow','lightpink'])
plt.title('Sales by Ship Mode')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#Average shipping day
sales['Order Date']=pd.to_datetime(sales['Order Date'], dayfirst=True,errors='coerce')
sales['Ship Date']=pd.to_datetime(sales['Ship Date'], dayfirst=True,errors='coerce')
sales['Shipping Days']=(sales['Ship Date']-sales['Order Date']).dt.days
round(sales['Shipping Days'].mean())
```

```python
#Top 10 Most Frequent Customer-Segment Combinations
sales.groupby('Customer Name')['Segment'].value_counts().sort_values(ascending=False).head(10)
```

```python
#Customer Lifetime Value (LTV) Analysis
sales['Order Date']=pd.to_datetime(sales['Order Date'], dayfirst=True,errors='coerce')
ltv=sales.groupby('Customer Name').agg({'Sales':'sum', 'Order Date':[min,max,'nunique']})
ltv.columns=['Total Sales', 'First Order', 'Last Order', 'Number of Orders']
ltv['Costumer Lifetime days']=(ltv['Last Order']-ltv['First Order']).dt.days+1
ltv['Avg Sales per day']=ltv['Total Sales']/ltv['Costumer Lifetime days']
ltv['LTV']=ltv['Avg Sales per day']*ltv['Number of Orders']
ltv=ltv.sort_values('Total Sales', ascending=False).head(10)
ltv
```

```python
#Top-10 most valuable customers based on Total Sales.
top_customers=sales.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_customers.plot(kind='bar', color='green')
plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#Conclusion

In the Sales distribution by year chart, we observe a positive trend in sales amount over the four years, along with a year-over-year increase in the number of orders. During these years, the highest sales occurred in the month of November. The month with the highest total sales was November, generating approximately 350,161.71 dollars, while the lowest sales were recorded in February, with around 59,371.12 dollars. Seasonal sales analysis indicates that the fall season consistently generated the highest sales across all observed years. The West and East regions recorded the highest total sales, with 710,219.68 dollars and 669,518.73 dollars respectively. The difference between them is approximately 1.8%, with the West leading slightly. California recorded the highest sales amount at 446,306.00 dollars, accounting for 19.73% of total orders, making it the leading state both in revenue and order frequency. New York City leads in both the number of orders and total sales, followed by Los Angeles. The Technology category generated the highest total sales at 827,455.90 dollars, accounting for 36.6% of overall sales. Furniture followed with 728,658.60 dollars (32.2%), and Office Supplies contributed 705,422.30 dollars, representing 31.2% of total sales. Binders and Paper are the top-performing sub-categories in terms of the number of orders. Phones and Chairs are the leading sub-categories by total sales, contributing 327,782.40 dollars(14.49%) and 322,822.70 dollars (14.27%) respectively. The Canon imageCLASS 2200 Advanced Copier leads in total sales by product name with 61 dollars ,599.82 from 5 units sold. In contrast, Staple Envelope and Staples have total sales of 1,675.65 dollars(47 units sold) and 755.47 dollars (46 units sold), respectively. The distribution of segments shows that Consumer accounts for 52.1%, Corporate for 30.1%, and Home Office for 17.8% of the total. The sales distribution by Ship Mode indicates that Standard Class is the most frequent, accounting for 59.3% of sales. The average shipping time is 4 days.
